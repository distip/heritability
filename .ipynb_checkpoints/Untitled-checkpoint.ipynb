```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('spectrum.csv')
```

```python
heritabilities = pd.read_csv('H_all.csv')
```

```python
data.head()
```

```python
data = data.loc[data['350'] > 0 , :]
```

```python
colnum=len(data.columns)
```

```python
plt.plot(data.iloc[ : , 5:colnum].T)
plt.xticks([148, 648, 1148, 1648], ['X500', 'X1000', 'X1500', 'X2000'])
plt.xlabel("Wavelenght (nm)")
plt.ylabel("Reflectance")
plt.title('Leaf Spectra of Sorghum Plants')
```

```python
plt.plot(heritabilities)
plt.xticks([150,650,1150,1650], ['X500','X1000','X1500','X2000'])#, rotation=90)
plt.xlabel("Wavelenght (nm)")
plt.ylabel("Heritability")
plt.title('Heritability of Spectral Data')
#plt.axvline(x=300, ymax=0.89)
#plt.axvline(x=590, ymax=0.85)
#plt.axvline(x=550, ymax=0.85)
#plt.axvline(x=650, ymax=0.85)
```

```python
fig, axs = plt.subplots(2, sharex=True, gridspec_kw={'height_ratios' : [1,3]})
axs[0].plot(heritabilities)
axs[0].set_ylabel('Heritability (%)')
axs[0].set_title('Broad Sense Heritability of Leaf Spectra', fontsize= 16)
plt.xticks([150,650,1150,1650], ['X500','X1000','X1500','X2000'])#, rotation=90)
axs[1].plot(data.iloc[ : , 5:colnum].T)
plt.ylabel('Reflectance', fontsize=13)
plt.xlabel('Wavelength (nm)', fontsize = 14)
fig.tight_layout()
```
